```python
import pandas as pd
previsores = pd.read_csv('entradas-breast.csv')
classe = pd.read_csv('saidas-breast.csv')
```

```python
from sklearn.model_selection import train_test_split
previsores_treinamento, previsores_teste, classe_treinamento, classe_test = train_test_split(previsores, classe, test_size = 0.25)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classificador = Sequential()

#primeira camada oculta
# número de neurônios na primeira camada oculta é igual:
# (nº de neurônios na camada de entrada + nº de neurônios na camada de saída)/2 
classificador.add(Dense(units = 16, activation = 'relu', kernel_initializer = 'random_uniform', input_dim = 30))
classificador.add(Dense(units = 16, activation = 'relu', kernel_initializer = 'random_uniform'))
classificador.add(Dense(units = 1, activation = 'sigmoid'))

#execução/criação da rede
optimizador = keras.optimizers.Adam(lr = 0.001, decay = 0.0001, clipvalue = 0.5)
classificador.compile(optimizer = optimizador, loss = 'binary_crossentropy', metrics = ['binary_accuracy'])



#treinamento da rede
classificador.fit(previsores_treinamento,classe_treinamento, batch_size=10, epochs=100)

#teste da rede
previsoes = classificador.predict(previsores_teste)
```

```python
pesos0 = classificador.layers[0].get_weights()
print(pesos0)
print(len(pesos0))
```

```python
previsoes = (previsoes>0.5)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
precisao = accuracy_score(classe_test, previsoes)
matrix = confusion_matrix(classe_test, previsoes)
print(matrix)
```

```python
resultado = classificador.evaluate(previsoes_teste, classe_test)
```
